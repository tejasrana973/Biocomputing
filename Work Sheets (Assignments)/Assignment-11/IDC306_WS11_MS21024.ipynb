```python
import re

entries = {}
entry = {}

with open('uniprotSnip.fasta', 'r') as file:
    for line in file:
        line = line.strip()
        if line.startswith('>'):
            if entry and 'UniqueIdentifier' in entry:
                entries[entry['UniqueIdentifier']] = entry
            entry = {}
            match = re.match(r'>(sp|tr)\|(\w+)\|(\w+).+OS=(.+)\sOX=(\d+)\s*(?:GN=(\w+))?\s*(?:PE=(\d+))?\s*(?:SV=(\d+))?$', line)
            if match:
                entry['UniqueIdentifier'] = match.group(2)
                entry['EntryName'] = match.group(3).split()[0]
                entry['OrganismName'] = match.group(4)
                entry['GeneName'] = match.group(6) if match.group(6) else None
                entry['ProteinExistence'] = match.group(7) if match.group(7) else None
            entry['ProteinSequence'] = ''
        else:
            if 'UniqueIdentifier' in entry:
                entry['ProteinSequence'] += line
                #print(entry)
        if entry and 'UniqueIdentifier' in entry:
            entries[entry['UniqueIdentifier']] = entry


for entry_id, entry_info in entries.items():
    print(f"\nID : {entry_id}")
    for key, value in entry_info.items():
        print(f"  {key}: {value}")
```

```python
filename = 'uniprotSnip.txt'
entries = []
entry = {}

with open(filename, 'r') as file:
    for line in file:
        line = line.strip()
        if line.startswith('ID'):
            if entry:
                entries.append(entry)
            entry = {'ID': re.search(r'ID\s+(\w+)', line).group(1)}
        elif line.startswith('AC'):
            entry['AC'] = re.search(r'AC\s+(.+?);', line).group(1)
        elif line.startswith('DR'):
            entry.setdefault('DR', []).extend(re.findall(r'DR\s+(.+)', line))
            #re.split(';', m.group(1))
        elif line.startswith('SQ'):
            sq_match = re.search(r'SQ\s+SEQUENCE\s+(\d+)\s+AA;\s+(\d+)\s+MW;.+', line)
            entry['SQ Length'] = sq_match.group(1)
            entry['SQ MW'] = sq_match.group(2)

if entry:
    entries.append(entry)

#print(entry)

for entry in entries:
    print("\nID:", entry['ID'])
    print("AC:", entry.get('AC'))
    print("DR:", entry.get('DR', []))
    print("SQ Length:", entry['SQ Length'])
    print("SQ MW:", entry['SQ MW'])
```
