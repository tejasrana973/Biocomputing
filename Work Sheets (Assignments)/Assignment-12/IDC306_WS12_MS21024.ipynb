```python
class Sequence:
    def __init__(self, sequence):
        self.sequence = sequence

    def composition(self):
        comp = {}
        for char in self.sequence:
            comp[char] = comp.get(char, 0) + 1
        return comp

    def is_dna(self):
        valid_dna_characters = {'A', 'T', 'C', 'G'}
        return all(char in valid_dna_characters for char in self.sequence)

    def is_protein(self):
        valid_protein_characters = set("ACDEFGHIKLMNPQRSTVWY")
        return all(char in valid_protein_characters for char in self.sequence)




# dna_seq = (input("Enter the dna sequence:").upper())
# protein_seq = (input("Enter the protein sequence:").upper())
dna_seq = "ATGGCTGACGCTGACGATGCTGATAATAGACAGCTGACGTGGATACAGATTAGCAGATACAGATTGACAGATACAG"
protein_seq = "ADCLAHGGSATTRWQCDGTRWFWQRGSYQWFSREDCWQGPLOLMNVGTDERWCS"

dna_sequence = Sequence(dna_seq)
protein_sequence = Sequence(protein_seq)

## Finding the composition of the Sequence
dna_composition = dna_sequence.composition()
print("DNA Composition:", dna_composition)

protein_composition = protein_sequence.composition()
print("Protein Composition:", protein_composition)


## Finding whether the input sequence is DNA/protein
print("\nFor DNA Sequence :\nIs DNA:", dna_sequence.is_dna())
print("Is Protein:", dna_sequence.is_protein())

print("\nFor Protein Sequence :\nIs DNA:", protein_sequence.is_dna())
print("Is Protein:", protein_sequence.is_protein())
```

```python
class Sequence:
    def __init__(self, sequence):
        self.sequence = sequence.upper()

    def find_longest_orf(self):
        """
        Find the longest Open Reading Frame (ORF) in the DNA sequence.
        """
        start_codon = ["ATG", "GTG"]
        stop_codons = ["TAA", "TAG", "TGA"]
        longest_orf = ""
        current_orf = ""
        all_orfs = []


        for i in range(len(self.sequence)):
            if self.sequence[i:i+3] in start_codon:

                current_orf = self.sequence[i:i+3]

                for j in range(i+3, len(self.sequence), 3):
                    codon = self.sequence[j:j+3]
                    if codon in stop_codons:
                        current_orf += codon
                        all_orfs.append(current_orf)

                        if len(current_orf) > len(longest_orf):
                            longest_orf = current_orf
                        current_orf = ""
                        break
                    else:
                      current_orf += codon

        return longest_orf, all_orfs


if __name__ == "__main__":
    dna_sequence = (input("Enter the DNA sequence:").upper())
    seq = Sequence(dna_sequence)
    longest_orf, all_orfs = seq.find_longest_orf()
    if longest_orf:
        print("Longest ORF:", longest_orf)
        print("All ORFs found:", all_orfs)
    else:
        print("No ORF found.")
```

```python
class FastaFile:
    def __init__(self, file_path):
        self.sequence_names = []
        self.fasta_sequences = {}
        self.read_fasta(file_path)

    def read_fasta(self, file_path):
        current_sequence_name = ""
        current_sequence = ""

        with open(file_path, 'r') as file:
            for line in file:
                line = line.strip()
                if line.startswith('>'):
                    if current_sequence_name:
                        self.sequence_names.append(current_sequence_name)
                        self.fasta_sequences[current_sequence_name] = current_sequence
                    current_sequence_name = line[1:]
                    current_sequence = ""
                else:
                    current_sequence += line

            if current_sequence_name:
                self.sequence_names.append(current_sequence_name)
                self.fasta_sequences[current_sequence_name] = current_sequence


fasta_file = FastaFile("fasta_file.txt")
print("Sequence names:", fasta_file.sequence_names)
print("Fasta sequences:")
for name, sequence in fasta_file.fasta_sequences.items():
    print(name, ":", sequence)
```
