```python
def Generate_data(file_path):
    data = {}
    f=open(file_path)
    for line in f.readlines():
      resid = line[24:27].strip()
      atom_name = line[12:16].strip()
      res_name = line[17:21].strip()
      x_coord = line[31:38].strip()
      y_coord = line[39:46].strip()
      z_coord = line[47:54].strip()

      if resid not in data:
        data[resid] = {'resname': res_name}

      if atom_name not in data[resid]:
        data[resid][atom_name] = [float(x_coord), float(y_coord), float(z_coord)]

    return data


PDB_data = Generate_data('atom_cord.txt')
print(PDB_data)
```

```python
import math

def calculate_angle(v1, v2, v3):
    n1 = [v1[1]*v2[2] - v1[2]*v2[1],
          v1[2]*v2[0] - v1[0]*v2[2],
          v1[0]*v2[1] - v1[1]*v2[0]]

    n2 = [v2[1]*v3[2] - v2[2]*v3[1],
          v2[2]*v3[0] - v2[0]*v3[2],
          v2[0]*v3[1] - v2[1]*v3[0]]

    n12 = [n1[1]*n2[2] - n1[2]*n2[1],
           n1[2]*n2[0] - n1[0]*n2[2],
           n1[0]*n2[1] - n1[1]*n2[0]]
    dot_product = sum(n1[i] * n2[i] for i in range(3))
    value = sum(n12[i] * v2[i] for i in range(3))

    sign = +1
    if value < 0:
        sign = sign * -1

    n1_magnitude = math.sqrt(sum(v**2 for v in n1))
    n2_magnitude = math.sqrt(sum(v**2 for v in n2))

    cosine_angle = dot_product / (n1_magnitude * n2_magnitude)
    angle = math.acos(cosine_angle)

    return math.degrees(angle) * sign

def phi(i):
  v1 = [PDB_data[i]['N'][j] - PDB_data[str(int(i)-1)]['C'][j] for j in range(3)]
  v2 = [PDB_data[i]['CA'][j] - PDB_data[i]['N'][j] for j in range(3)]
  v3 = [PDB_data[i]['C'][j] - PDB_data[i]['CA'][j] for j in range(3)]
  phi = calculate_angle(v1,v2,v3)

  return phi

def psi(i):
  v1 = [PDB_data[i]['CA'][j] - PDB_data[i]['N'][j] for j in range(3)]
  v2 = [PDB_data[i]['C'][j] - PDB_data[i]['CA'][j] for j in range(3)]
  v3 = [PDB_data[str(int(i)+1)]['N'][j] - PDB_data[i]['C'][j] for j in range(3)]
  psi = calculate_angle(v1,v2,v3)

  return psi

Res_id = str(int(input("Enter Residue ID : ")))
print(f"\nFor Residue {Res_id},\nPhi Angle : {phi(Res_id)} degrees \nPsi Angle : {psi(Res_id)} degrees")
```
