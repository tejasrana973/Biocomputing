```python
def ReadPDB(file_path):
    data = {}
    f=open(file_path)
    for line in f.readlines():
      resid = line[23:27].strip()
      atom_name = line[12:16].strip()
      res_name = line[17:21].strip()
      x_coord = line[31:38].strip()
      y_coord = line[39:46].strip()
      z_coord = line[47:54].strip()

      if resid not in data:
        data[resid] = {'resname': res_name}

      if atom_name not in data[resid]:
        data[resid][atom_name] = [x_coord, y_coord, z_coord]

    return data


PDB_data = ReadPDB('atom_cord.txt')
print(f"The required data in dictionary form is :\n{PDB_data}")
```

```python
def Find_Composition(pdb_data,AA):
  comp = {}
  for resid, data in pdb_data.items():
    resname = data['resname']
    comp[resname] = comp.get(resname,0) + 1

  A_count = comp[AA]

  return comp,A_count

Amino_acid = input("Enter an amino acid : ").upper()
Composition,freqAA = Find_Composition(PDB_data,Amino_acid)
print(f"The frequency of {Amino_acid} is : {freqAA}")
print(f"\nThe composition of all residues is :\n{Composition}")
```

```python
import math

def calculate_distance(v1, v2):
    v_diff = [float(v2[i]) - float(v1[i]) for i in range(3)]
    return math.sqrt(sum(v**2 for v in v_diff))

def Find_Dist(pdb_data,resid1,resid2,atom1,atom2):
  v1 = pdb_data[str(resid1)][atom1]
  v2 = pdb_data[str(resid2)][atom2]
  d = calculate_distance(v1,v2)
  return d

(Resid_1,Atom_1) = (5,'N')
(Resid_2,Atom_2) = (7,'CA')
Dist = Find_Dist(PDB_data,Resid_1,Resid_2,Atom_1,Atom_2)
print(f"The distance between Residue-{Resid_1} Atom {Atom_1} and Residue-{Resid_2} Atom {Atom_2} is : {Dist}")
```

```python
import math

def calculate_angle(atom1, atom2, atom3):
    vector_a = [float(atom1[i]) - float(atom2[i]) for i in range(3)]
    vector_b = [float(atom3[i]) - float(atom2[i]) for i in range(3)]

    dot_product = sum(vector_a[i] * vector_b[i] for i in range(3))

    magnitude_a = math.sqrt(sum(v**2 for v in vector_a))
    magnitude_b = math.sqrt(sum(v**2 for v in vector_b))

    cosine_angle = dot_product / (magnitude_a * magnitude_b)

    angle_radians = math.acos(cosine_angle)

    angle_degrees = math.degrees(angle_radians)

    return angle_degrees

def Find_Angle(pdb_data,r1,r2,r3,a1,a2,a3):
  v1 = pdb_data[str(r1)][a1]
  v2 = pdb_data[str(r2)][a2]
  v3 = pdb_data[str(r3)][a3]
  angle = calculate_angle(v1,v2,v3)
  return angle

(Resid_1,Atom_1) = (5,'N')
(Resid_2,Atom_2) = (7,'CA')
(Resid_3,Atom_3) = (10,'O')
Angle = Find_Angle(PDB_data,Resid_1,Resid_2,Resid_3,Atom_1,Atom_2,Atom_3)
print(f"The angle between {Atom_1}-{Atom_2}-{Atom_3} of residues {Resid_1},{Resid_2},{Resid_3} respectively is : {Angle} Degrees")
```
